```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data ingestion
path = '/content/drive/MyDrive/Causes of Death/causes-of-death-in-children-under-5 NEW.csv'
df = pd.read_csv(path)
```

```python
#beginning of data validation
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
#converting Year to a string for correlation analysis, as I do not want to correlate the year as a numerical value
df['Year']=df['Year'].astype('str')
```

```python
#adjusting data types to return Entity and Code as strings
df = df.convert_dtypes()
```

```python
df.dtypes
```

```python
#checking for nulls
df.isna().sum()
```

```python
#filtering for regional data
region_df = df[df['Code']=='0']
region_df
```

```python
#it looks like that retained more than expected, individual countries and data from two sources still present here
region_df.Entity.unique()
```

```python
#creating a dataframe of only World Bank regional data for viz
wb_region_df = region_df[region_df['Entity'].str.contains('WB')]
wb_region_df
```

```python
#defining function for Z_score
def calc_z(df, col):
  mean = df[col].mean()
  std = df[col].std()
  return (df[col]-mean)/std
```

```python
#checking for outliers based on Z-score, looks like none
dd_z = calc_z(wb_region_df,'Deaths - Digestive diseases - Sex: Both - Age: Under 5 (Number)')
dd_z[dd_z.abs() >= 3]
```

```python
#setting up Pearson's correlation analysis for all World Bank Regions
corr=wb_region_df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(data=corr,
            annot=True,
            cmap='RdBu',
            vmin=-1);
```

```python
#creating a dataframe of only World Health Organization regional data for viz
who_region_df = region_df[region_df['Entity'].str.contains('WHO')]
who_region_df
```

```python
#checking WHO data
who_region_df['Entity'].values.unique()
```

```python
entity_df = df.groupby('Entity')
```

```python
#making subset World Health Organization dataframes for viz of each region
ar_who_df = who_region_df[who_region_df['Entity'].str.contains('African')]
er_who_df = who_region_df[who_region_df['Entity'].str.contains('European')]
sear_who_df = who_region_df[who_region_df['Entity'].str.contains('South')]
emr_who_df = who_region_df[who_region_df['Entity'].str.contains('Eastern')]
ra_who_df = who_region_df[who_region_df['Entity'].str.contains('Americas')]
wpr_who_df = who_region_df[who_region_df['Entity'].str.contains('Pacific')]
```

```python
#making subset World Bank dataframes for viz of each region
na_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('North America')]
eap_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('East Asia')]
eca_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('Europe')]
lac_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('Latin')]
mena_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('Middle East')]
sa_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('South Asia')]
ssa_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('Sub-Saharan')]
```

```python
#visualizing deaths over time due to Digestive diseases in all World Health Organization regions
plt.figure(figsize=(12,8))
plt.plot(ar_who_df['Year'],ar_who_df['Deaths - Digestive diseases - Sex: Both - Age: Under 5 (Number)'],label='African Region')
plt.plot(er_who_df['Year'],er_who_df['Deaths - Digestive diseases - Sex: Both - Age: Under 5 (Number)'],label='European Region')
plt.plot(sear_who_df['Year'],sear_who_df['Deaths - Digestive diseases - Sex: Both - Age: Under 5 (Number)'],label='Southeast Asia Region')
plt.plot(emr_who_df['Year'],emr_who_df['Deaths - Digestive diseases - Sex: Both - Age: Under 5 (Number)'],label='Eastern Mediterranean Region')
plt.plot(ra_who_df['Year'],ra_who_df['Deaths - Digestive diseases - Sex: Both - Age: Under 5 (Number)'],label='Region of the Americas')
plt.plot(wpr_who_df['Year'],wpr_who_df['Deaths - Digestive diseases - Sex: Both - Age: Under 5 (Number)'],label='Western Pacific Region')
plt.title('Global Deaths in Children Under 5 due to Digestive Diseases (WHO)',fontsize=20)
plt.ylabel('Number of Deaths',fontsize=15)
plt.xlabel('Year',fontsize=15)
plt.xticks(rotation=90)
plt.legend();
```
