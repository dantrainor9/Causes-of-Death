```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data ingestion
path = '/content/drive/MyDrive/Causes of Death/causes-of-death-in-children-under-5 NEW.csv'
df = pd.read_csv(path)
```

```python
#adjusting data types to return Entity and Code as strings
df = df.convert_dtypes()
```

```python
#filtering for regional data
region_df = df[df['Code']=='0']
region_df
```

```python
region_df.Entity.unique()
```

```python
#creating a dataframe of only World Bank regional data for viz
wb_region_df = region_df[region_df['Entity'].str.contains('WB')]
wb_region_df
```

```python
#making subset dataframes for viz of each region
na_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('North America')]
eap_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('East Asia')]
eca_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('Europe')]
lac_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('Latin')]
mena_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('Middle East')]
sa_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('South Asia')]
ssa_wb_df = wb_region_df[wb_region_df['Entity'].str.contains('Sub-Saharan')]
```

```python
#visualizing deaths over time due to Nutritional deficiencies in all World Bank regions
plt.figure(figsize=(12,8))
plt.plot(na_wb_df['Year'],na_wb_df['Deaths - Nutritional deficiencies - Sex: Both - Age: Under 5 (Number)'],label='North America')
plt.plot(eap_wb_df['Year'],eap_wb_df['Deaths - Nutritional deficiencies - Sex: Both - Age: Under 5 (Number)'],label='East Asia & Pacific')
plt.plot(eca_wb_df['Year'],eca_wb_df['Deaths - Nutritional deficiencies - Sex: Both - Age: Under 5 (Number)'],label='Europe & Central Asia')
plt.plot(lac_wb_df['Year'],lac_wb_df['Deaths - Nutritional deficiencies - Sex: Both - Age: Under 5 (Number)'],label='Latin America & Carribean')
plt.plot(mena_wb_df['Year'],mena_wb_df['Deaths - Nutritional deficiencies - Sex: Both - Age: Under 5 (Number)'],label='Middle East & North Africa')
plt.plot(sa_wb_df['Year'],sa_wb_df['Deaths - Nutritional deficiencies - Sex: Both - Age: Under 5 (Number)'],label='South Asia')
plt.plot(ssa_wb_df['Year'],ssa_wb_df['Deaths - Nutritional deficiencies - Sex: Both - Age: Under 5 (Number)'],label='Sub-Saharan Africa')
plt.title('Global Deaths in Children Under 5 due to Nutritional Deficiencies',fontsize=20)
plt.ylabel('Number of Deaths',fontsize=15)
plt.xlabel('Year',fontsize=15)
plt.legend();
```
